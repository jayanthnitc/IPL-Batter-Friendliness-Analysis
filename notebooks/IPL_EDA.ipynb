```python
innings_scores = (
    ipl.groupby(
        ['season', 'match_id', 'inning']
    )['total_runs']
    .sum()
    .reset_index()
)
```

```python
innings_scores.head()
```

```python
avg_score_by_season = (
    innings_scores.groupby('season')['total_runs']
    .mean()
    .reset_index()
)
```

```python
avg_score_by_season.columns = [
    'season',
    'avg_innings_score'
]
```

```python
avg_score_by_season.head()
```

```python
innings_scores['score_200_plus'] = (
    innings_scores['total_runs'] >= 200
)
```

```python
innings_scores.head()
```

```python
scores_200 = (
    innings_scores.groupby('season')
    ['score_200_plus']
    .sum()
    .reset_index()
)
```

```python
scores_200.columns = [
    'season',
    'count_200_plus'
]
```

```python
ipl['is_four'] = (
    ipl['batsman_runs'] == 4
)
```

```python
ipl['is_six'] = (
    ipl['batsman_runs'] == 6
)
```

```python
boundary_stats = (
    ipl.groupby('season')
    [['is_four', 'is_six']]
    .sum()
    .reset_index()
)
```

```python
boundary_stats.head()
```

```python
season_runs = (
    ipl.groupby('season')['total_runs']
    .sum()
)
```

```python
season_balls = (
    ipl.groupby('season')
    .size()
)
```

```python
run_rate = (
    season_runs / season_balls
) * 6
```

```python
run_rate = run_rate.reset_index()
run_rate.columns = [
    'season',
    'run_rate'
]
```

```python
economy_by_season = (
    ipl.groupby('season')['total_runs']
    .sum()
    /
    (ipl.groupby('season').size() / 6)
).reset_index()
```

```python
economy_by_season.columns = [
    'season',
    'economy_rate'
]
```

```python
economy_by_season.head()
```

```python
plt.figure(figsize=(12,5))

plt.plot(
    economy_by_season['season'],
    economy_by_season['economy_rate'],
    marker='o'
)

plt.title('Season-wise Economy Rate')
plt.xlabel('Season')
plt.ylabel('Economy Rate')
plt.grid(True)

plt.show()
```

```python
##venue analysis
venue_scores = (
    ipl.groupby(
        ['venue', 'match_id', 'inning']
    )['total_runs']
    .sum()
    .reset_index()
)
```

```python
venue_avg = (
    venue_scores.groupby('venue')['total_runs']
    .mean()
    .sort_values(ascending=False)
    .reset_index()
)

venue_avg.columns = [
    'venue',
    'avg_score'
]
```

# Batter Friendliness Index (BFI)

To quantify batting-friendliness across IPL seasons, a custom Batter Friendliness Index (BFI) was developed.

The index combines:
- Average Innings Score
- Run Rate
- Six-Hitting Frequency
- Bowling Economy Rate

Higher BFI values indicate more batting-friendly conditions.

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
bfi_df = bfi.copy()
```

```python
scaler = MinMaxScaler()

bfi_df[
    [
        'avg_innings_score',
        'run_rate',
        'is_six',
        'economy_rate'
    ]
] = scaler.fit_transform(
    bfi_df[
        [
            'avg_innings_score',
            'run_rate',
            'is_six',
            'economy_rate'
        ]
    ]
)
```

```python
bfi_df['BFI'] = (
    bfi_df['avg_innings_score']
    + bfi_df['run_rate']
    + bfi_df['is_six']
    + bfi_df['economy_rate']
) / 4
```

```python
bfi_df['BFI'] = bfi_df['BFI'] * 100
```

```python
bfi_df[['season', 'BFI']].head()
```

```python
plt.figure(figsize=(12,6))

plt.plot(
    bfi_df['season'],
    bfi_df['BFI'],
    marker='o',
    linewidth=2
)

plt.title('Batter Friendliness Index (BFI) by Season')
plt.xlabel('Season')
plt.ylabel('BFI Score')
plt.grid(True)

plt.show()
```

## Observation

The Batter Friendliness Index (BFI) demonstrates a clear increase over time.

Recent IPL seasons exhibit significantly higher batting-friendliness compared to the early seasons, driven by higher scoring rates, increased six-hitting, and elevated bowling economy rates.
