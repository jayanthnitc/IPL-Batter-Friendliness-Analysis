```python
innings_scores = (
    ipl.groupby(
        ['season', 'match_id', 'inning']
    )['total_runs']
    .sum()
    .reset_index()
)
```

```python
innings_scores.head()
```

```python
avg_score_by_season = (
    innings_scores.groupby('season')['total_runs']
    .mean()
    .reset_index()
)
```

```python
avg_score_by_season.columns = [
    'season',
    'avg_innings_score'
]
```

```python
avg_score_by_season.head()
```

```python
innings_scores['score_200_plus'] = (
    innings_scores['total_runs'] >= 200
)
```

```python
innings_scores.head()
```

```python
scores_200 = (
    innings_scores.groupby('season')
    ['score_200_plus']
    .sum()
    .reset_index()
)
```

```python
scores_200.columns = [
    'season',
    'count_200_plus'
]
```

```python
ipl['is_four'] = (
    ipl['batsman_runs'] == 4
)
```

```python
ipl['is_six'] = (
    ipl['batsman_runs'] == 6
)
```

```python
boundary_stats = (
    ipl.groupby('season')
    [['is_four', 'is_six']]
    .sum()
    .reset_index()
)
```

```python
boundary_stats.head()
```

```python
season_runs = (
    ipl.groupby('season')['total_runs']
    .sum()
)
```

```python
season_balls = (
    ipl.groupby('season')
    .size()
)
```

```python
run_rate = (
    season_runs / season_balls
) * 6
```

```python
run_rate = run_rate.reset_index()
run_rate.columns = [
    'season',
    'run_rate'
]
```
